```python
import random
import re
from collections import defaultdict

class MarkovTextGenerator:
    def __init__(self, order=2):
        self.order = order
        self.model = defaultdict(list)

    def clean_text(self, text):
        text = text.lower()
        text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
        return text.split()

    def train(self, text):
        words = self.clean_text(text)
        for i in range(len(words) - self.order):
            key = tuple(words[i:i + self.order])
            next_word = words[i + self.order]
            self.model[key].append(next_word)

    def generate_text(self, length=40):
        start = random.choice(list(self.model.keys()))
        output = list(start)

        for _ in range(length):
            state = tuple(output[-self.order:])
            if state not in self.model:
                break
            output.append(random.choice(self.model[state]))

        return " ".join(output)
```

```python
sample_text = """
Artificial intelligence is transforming the world.
Machine learning allows systems to learn from data.
Markov chains are simple but powerful models.
Text generation is a core concept in natural language processing.
"""

generator = MarkovTextGenerator(order=2)
generator.train(sample_text)

print("Generated Text:\n")
print(generator.generate_text())
```
